```python
import os
def create_artifacts():
    path = os.path.abspath('data_cleaning.ipynb')
    index = path.find('datacollection')
    dir_path = path[:index]+'dashboard\\cleanedData\\'
    print(dir_path)
    
    if not os.path.exists(dir_path):
        os.makedirs(dir_path)
        
    match_df.to_csv(dir_path+'match_result.csv')
    new_df.to_csv(dir_path+'batting_summary.csv')
    bowling_df1.to_csv(dir_path+'bowling_summary.csv')
    player_df1.to_csv(dir_path+'player_details.csv')
```

```python
create_artifacts()
```
