### COLLECTING BATTING AND BOWLING DATA

```python
import re
dict1 = {}
links = soup.find_all('a', attrs={"class" : "data-link", "href":re.compile("^/ci/engine/match")})
for link in links:
    string = link.string
    base_url = 'https://stats.espncricinfo.com'
    pg_content = requests.get(base_url+link.get('href'))
    dict1[string] = pg_content
    print(pg_content)
    print(link.get('href'))
```

```python
dict1
```

```python
df_batting = pd.DataFrame()
df_bowling = pd.DataFrame()

for id, response in dict1.items():
    soup2 = BeautifulSoup(response.content, 'html.parser')
    teams = soup2.find_all('span', {'class':'ds-text-tight-l ds-font-bold ds-text-ui-typo hover:ds-text-ui-typo-primary ds-block ds-truncate'})
    team_a = teams[0].string
    team_b = teams[1].string
    match = team_a + " VS " + team_b
    team_list = [team_a,team_b]
    
    batting_table = soup2.find_all('table', {"class":"ds-w-full ds-table ds-table-md ds-table-auto ci-scorecard-table"})
    bowling_table = soup2.find_all('table', {"class":"ds-w-full ds-table ds-table-md ds-table-auto"})
    
    for index, table in enumerate(batting_table):
        df_bat = pd.read_html(str(table))[0]
        df_bat.insert(loc=0, column='match', value = match )
        df_bat.insert(loc=1, column='team_innings', value = team_list[index] )
        df_bat.insert(loc=2, column='batting_pos', value = range(df_bat.shape[0]) )
        df_bat.insert(loc=len(df_bat.columns), column='match_id', value = id )
        df_batting = pd.concat([df_batting,df_bat],ignore_index=True)
    
    for index,table in enumerate(balling_table):
        df_bwl = pd.read_html(str(table))[0]
        df_bwl.insert(loc=0, column='match', value = match )
        df_bwl.insert(loc=1, column='bowling_team', value = team_list[index] )
        df_bwl.insert(loc=len(df_bwl.columns), column='match_id', value = id )
        df_bowling = pd.concat([df_bowling,df_bwl],ignore_index=True)

display(df_batting)
display(df_bowling)
```

```python
df_batting.to_csv('batting_summary.csv')
df_balling.to_csv('bowlng_summary.csv')
```

### COLLECTING PLAYER DETAILS DATA

```python
player_link_list = []
for id, response in dict1.items():
    soup3 = BeautifulSoup(response.content, 'html.parser')
    a = soup3.find_all('a', {'href':re.compile("^/player/")})
    for i in a:
       player_link_list.append(i.get('href'))
```

```python
unique_players = list(set(player_link_list))
```

```python
from selenium import webdriver
player_df = pd.DataFrame()
base_url = 'https://stats.espncricinfo.com'
for player in unique_players:
    browser = webdriver.Chrome()
    browser.get(base_url+player)
    res = browser.page_source
    player_soup = BeautifulSoup(res, 'html.parser')
    try:
        for i in player_soup.find_all('div',{'class':"ds-ml-auto ds-w-48 ds-h-48"}):
            img_link = i.find('img').get('src')
    except:
        img_link = ""
    player_div = player_soup.find('div', attrs={'class':'ds-p-4'})
    name = player_soup.find('h1', {'class':'ds-text-title-l ds-font-bold'}).text
    headers = player_div.find_all('p',attrs={'class':'ds-text-tight-m ds-font-regular ds-uppercase ds-text-ui-typo-mid'})
    headers = [i.text for i in headers]
    headers.insert(0,'Name')
    headers.append('Team')
    headers.append('image_link')
    values = player_div.find_all('span',attrs={'class':'ds-text-title-s ds-font-bold ds-text-ui-typo'})
    values = [i.text for i in values]
    try:
        player_team = player_div.find('div', {"class":"ds-grid lg:ds-grid-cols-3 ds-grid-cols-2 ds-gap-y-4"}).find('a').text
    except:
        player_team = ""
    name = player_soup.find('h1',{'class':'ds-text-title-l ds-font-bold'}).text
    values.insert(0,name)
    values.append(player_team)
    values.append(img_link)
    player_details_dict = dict(zip(headers,values))
    player_dff = pd.DataFrame(player_details_dict, index=[0])
    player_df = pd.concat([player_df,player_dff],ignore_index=True)
    browser.quit()
```

```python
player_df.to_csv('player_details.csv')
```
